```python
import fbprophet
print(fbprophet.__version__)
```

```python
import pandas as pd

df = pd.read_csv('naver_keyword.csv')
df.head()
```

```python
df.columns = ['ds', 'y']
df['ds'] = pd.to_datetime(df['ds'])

from fbprophet import Prophet
model = Prophet()

# 모델 학습
model.fit(df)
```

```python
import datetime
from dateutil.relativedelta import relativedelta 


start = df['ds'].iloc[-1]
end = df['ds'].iloc[-1] - relativedelta(months=-1)
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]


date_list = []
for date in date_generated:
    date_list.append(date.strftime("%Y-%m-%d"))
    
    
date_list = pd.DataFrame(date_list, columns = ['ds'])
date_list['ds']= pd.to_datetime(date_list['ds']) 
date_list
```

```python
forecast = model.predict(date_list)
```

```python
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```

```python
predict_data = forecast[['ds','yhat']]
predict_data.to_csv('predict_nsearch.csv', index=False)
```
